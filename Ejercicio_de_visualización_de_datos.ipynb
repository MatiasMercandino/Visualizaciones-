```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


filename = "/content/Global import of cocoa beans.xlsx"
df = pd.read_excel(filename)
df.head()
```

```python
df["Region"].value_counts()
df.info()
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df.head()
```

Filtrar por Region


```python
Region1=(df["Region"]=="Asia & Oceania")
Region1=df[Region1]
Region1
```

```python
Region2=(df["Region"]=="Europe")
Region2=df[Region2]
Region2
```

```python
Region3=(df["Region"]=="Americas")
Region3=df[Region3]
Region3
```

Importaciones por a침o de cada region

```python
plt.style.use('fivethirtyeight')
plt.plot(Region1["Year"],Region1["Import in tonnes"],c="magenta",marker=".",markersize=20,label = 'Region1');
plt.plot(Region2["Year"],Region2["Import in tonnes"],c="red",marker="p",markersize=20, label = 'Region2');
plt.plot(Region3["Year"],Region3["Import in tonnes"],c="blue",marker="*",markersize=20, label = 'Region3');
plt.xlabel("A침o")
plt.ylabel("Importaciones en [Ton]")
plt.title("Importaciones por region por a침o en toneladas ") 
plt.legend()
plt.legend(loc=(1.02,0))
plt.xticks(fontsize = 15);
plt.yticks(fontsize = 10);
plt.savefig('MATLABSavedPlot.png', dpi = 300)
plt.xticks(rotation = 90);
```

Importaciones por A침o

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,8))
# Subplot 1 
plt.subplot(1, 3, 1);
plt.plot(Region1["Year"], Region1["Import in tonnes"], c = 'r',marker=".",markersize=20,label="Asia & Oceania");
plt.xticks(rotation = 90);
plt.ylabel("Importaciones [Toneladas]",fontsize=15)
plt.title("Importaciones por A침o", fontsize=15)
plt.grid()
plt.legend()

#plt.scatter(df_intercept['feature'].values, df_intercept['actual'].values, c= 'k');
#plt.title('intercept', fontsize = 12);
# Subplot 2
plt.subplot(1, 3, 2);
plt.plot(Region2["Year"], Region2["Import in tonnes"], c = 'b',marker="p",markersize=20,label="Europe");
plt.xticks(rotation = 90);
plt.ylabel("Importaciones [Toneladas]",fontsize=15)
plt.title("Importaciones por A침o", fontsize=15)
plt.grid()
plt.legend()

#plt.scatter(df_nointercept['feature'].values, df_nointercept['actual'].values, c= 'k');
#plt.title('no intercept', fontsize = 12);
plt.subplot(1, 3, 3);
plt.plot(Region3["Year"], Region3["Import in tonnes"], c = 'y',marker="*",markersize=20,label="Americas");
plt.xticks(rotation = 90);
plt.ylabel("Importaciones [Toneladas]",fontsize=15)
plt.title("Importaciones por A침o", fontsize=15)
plt.grid()
plt.legend()

plt.tight_layout()
```
